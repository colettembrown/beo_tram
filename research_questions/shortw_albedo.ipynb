```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import glob
import os
import seaborn as sns
```

```python
tram_data = pd.read_csv("../data/tram_complete.csv")
```

```python
tram_2014 = tram_data[(tram_data['Year'] == 2014) & (tram_data['date'] <= '2014-07-30')]
tram_2015 = tram_data[(tram_data['Year'] == 2015) & (tram_data['date'] <= '2015-07-30')]
tram_2016 = tram_data[(tram_data['Year'] == 2016) & (tram_data['date'] <= '2016-07-30')]
tram_2017 = tram_data[(tram_data['Year'] == 2017) & (tram_data['date'] <= '2017-07-30')]
```

```python
tram_2014_daily_mean = tram_2014.groupby('date')[['ShortW_inc', 'Albedo']].mean().reset_index()

fig, ax1 = plt.subplots(figsize=(10, 6))

# Plot ShortW_inc on the left y-axis
ax1.plot(tram_2014_daily_mean['date'], tram_2014_daily_mean['ShortW_inc'], label='Shortwave Inc.', color='blue')
ax1.set_xlabel('Date')
ax1.set_ylabel('Shortwave Inc.', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')

# Create a new axis for Albedo on the right y-axis
ax2 = ax1.twinx()
ax2.plot(tram_2014_daily_mean['date'], tram_2014_daily_mean['Albedo'], label='Albedo', color='orange')
ax2.set_ylabel('Albedo', color='orange')
ax2.tick_params(axis='y', labelcolor='orange')

# Add title and format x-axis
plt.title('Shortwave Inc. and Albedo over Time')
plt.xticks(rotation=90)
plt.tight_layout()

plt.show()
```
