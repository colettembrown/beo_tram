```python
import pandas as pd
```

```python
tram_2014 = pd.read_csv("/Users/colettebrown/repos/tram/data/NGEE-Barrow_tram_final_2014_data.csv")
tram_2015 = pd.read_csv("/Users/colettebrown/repos/tram/data/NGEE-Barrow_tram_final_2015_data.csv")
tram_2016 = pd.read_csv("/Users/colettebrown/repos/tram/data/NGEE-Barrow_tram_final_2016_data.csv")
tram_2017 = pd.read_csv("/Users/colettebrown/repos/tram/data/NGEE-Barrow_tram_final_2017_data.csv")
```

```python
tram_data = pd.concat([tram_2014, tram_2015, tram_2016, tram_2017])
```

```python
tram_data.columns
```

```python
len(tram_data[tram_data['Year '] == 2016])
```

```python
tram_data = tram_data[tram_data['Albedo'] != -9999]
tram_data = tram_data[tram_data['Albedo'] <= 0.95]
tram_data = tram_data[tram_data['Albedo'] >= 0.01]
tram_data = tram_data[tram_data['ShortW_inc'] >= 50]
tram_data = tram_data[tram_data['NDVI'] != -9999]
```

```python
len(tram_data[tram_data['Year '] == 2016])
```

```python
tram_data.columns
```

```python
tram_data.columns = tram_data.columns.str.strip()

tram_data.rename(columns={'Day_of_Month' : 'Day'}, inplace=True)

tram_data['date'] = pd.to_datetime(tram_data[['Year', 'Month', 'Day']])
```

```python
tram_data.describe()
```

```python
stop_metadata = pd.read_csv('/Users/colettebrown/repos/tram/data/Stop_Metadata.csv')
```

```python
stop_metadata.head()
```

```python
stop_metadata.drop(columns=['Distance_along_Tram', 'Elevation_of_Position', "Polygon_Type_Feature_ID", 'Unnamed: 6', 'Unnamed: 7'], inplace=True)
stop_metadata.dropna(inplace=True)
```

```python
stop_metadata['Polygon_Type'].unique()
```

```python
stop_metadata.rename(columns={'Position_for_Measurement':'Stop_number', 'Polygon_Type':'Polygon_type', 'Polygon_Feature':'Topo_feature'}, inplace=True)
stop_metadata
```

```python
tram_data = tram_data.merge(stop_metadata, on ='Stop_number')
```
