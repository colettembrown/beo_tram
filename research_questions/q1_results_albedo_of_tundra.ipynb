```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import glob
import os
import seaborn as sns
```

```python
tram_data = pd.read_csv("../data/tram_complete.csv")
```

```python
tram_2014 = tram_data[tram_data['Year'] == 2014]
tram_2015 = tram_data[tram_data['Year'] == 2015]
tram_2016 = tram_data[tram_data['Year'] == 2016]
tram_2017 = tram_data[tram_data['Year'] == 2017]
```

```python
tram_2014_daily_mean = tram_2014.groupby('date')[['ShortW_inc', 'Albedo']].mean().reset_index()

fig, ax1 = plt.subplots(figsize=(10, 6))

# Plot ShortW_inc on the left y-axis
ax1.plot(tram_2014_daily_mean['date'], tram_2014_daily_mean['ShortW_inc'], label='Shortwave Inc.', color='blue')
ax1.set_xlabel('Date')
ax1.set_ylabel('Shortwave Inc.', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')

# Create a new axis for Albedo on the right y-axis
ax2 = ax1.twinx()
ax2.plot(tram_2014_daily_mean['date'], tram_2014_daily_mean['Albedo'], label='Albedo', color='orange')
ax2.set_ylabel('Albedo', color='orange')
ax2.tick_params(axis='y', labelcolor='orange')

# Add title and format x-axis
plt.title('Shortwave Inc. and Albedo over Time')
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```

# Projected Albedo

```python
path_to_projections = "/Users/colettebrown/repos/tram/data/projections/"
```

```python
group_order = ['_spring_snow', '_spring_transition', '_peak_NDVI']
variables = ['Albedo', 'NDVI', 'fPAR', 'Net_rad']

polygon_types = {
    1: 'Trough & LCP',
    2: 'Trough & FCP',
    3: 'Trough & HCP',
    4: 'Center & LCP',
    5: 'Center & FCP',
    7: 'High & LCP',
    8: 'High & FCP',
    9: 'High & HCP'
}
```

```python
# Read in the projected data
projected_data = pd.DataFrame()

for group in group_order:
    for variable in variables:
        file_pattern = os.path.join(path_to_projections, f"{variable}{group}.txt")
        for file in glob.glob(file_pattern):
            file_name = os.path.basename(file)
            
            # Extract the group and variable from the file name
            df = pd.read_csv(file, sep='\t')
            df['Group'] = group
            df['Variable'] = variable
            
            projected_data = pd.concat([projected_data, df], ignore_index=True)

# Clean up the data
projected_data.replace(polygon_types, inplace=True)
```

### Plotting Mean Albedo by Polygon Type

```python
# Calculate mean and standard deviation for each group
group_order = ['_spring_snow', '_spring_transition', '_peak_NDVI']
grouped_stats = projected_data.groupby(['e00', 'Group', 'Variable'], as_index=False)['Var4'].agg(['mean', 'std']).reset_index()
grouped_stats = grouped_stats[grouped_stats['Variable'] == 'Albedo']
grouped_stats['Group'] = pd.Categorical(grouped_stats['Group'], categories=group_order, ordered=True)

# Sort by the categorical order
grouped_stats = grouped_stats.sort_values('Group')

# Plotting mean with shaded standard deviation
plt.figure(figsize=(10, 6))
for topo_feature, data in grouped_stats.groupby('e00'):
    plt.plot(data['Group'], data['mean'], marker='o', label=topo_feature)
    plt.fill_between(data['Group'], data['mean'] - data['std'], data['mean'] + data['std'], alpha=0.2)

plt.title('Albedo Changes Across Seasons for Each Topo Feature')
plt.xlabel('Season')
plt.ylabel('Average Albedo')
plt.xticks(rotation=45)
plt.legend(title='Topo Feature')
plt.show()
```

```python
heatmap_data = grouped_stats.pivot(index='e00', columns='Group', values='mean')
sns.heatmap(heatmap_data, annot=True, cmap='YlGnBu', fmt=".2f", cbar_kws={'label': 'Mean Albedo'})
plt.title("Mean Albedo Across Seasons and Topographic Features")
plt.xlabel("Time Period")
plt.ylabel("Polygon Type and Topographic Feature")
plt.show()
```

```python
avg_per_season = projected_data.loc[(projected_data['e00'] != 'High & HCP') & (projected_data['e00'] != 'Trough & HCP')]
seasonally_grouped = avg_per_season.groupby(['Group', 'Variable'], as_index=False)['Var4'].mean()
annual_albedo_projected = seasonally_grouped[seasonally_grouped['Variable'] == 'Albedo']
annual_albedo_projected['Group'] = pd.Categorical(annual_albedo_projected['Group'], categories=group_order, ordered=True)
```
